```python
# ============================================
# setup.ipynb - Data loading, exploration, and verification
# ============================================

# 1️⃣ Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import platform
import tensorflow as tf
import sklearn

# 2️⃣ Verify Environment
print("✅ Environment Check")
print(f"Python version: {platform.python_version()}")
print(f"Pandas version: {pd.__version__}")
print(f"Numpy version: {np.__version__}")
print(f"TensorFlow version: {tf.__version__}")
print(f"Scikit-learn version: {sklearn.__version__}")

# 3️⃣ Load Metadata
metadata_path = "/content/metadata.csv"
df = pd.read_csv(metadata_path)

print("\n✅ Dataset Loaded Successfully!")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")

# Display first few rows
display(df.head())

# 4️⃣ Basic Dataset Information
print("\n📊 Dataset Info:")
df.info()

print("\n🔍 Checking for Missing Values:")
print(df.isnull().sum())

# 5️⃣ Summary Statistics
print("\n📈 Summary Statistics:")
display(df.describe(include='all'))

# 6️⃣ Explore Class Distribution
if 'diagnostic' in df.columns:
    plt.figure(figsize=(8,5))
    sns.countplot(y='diagnostic', data=df, palette='viridis')
    plt.title("Class Distribution")
    plt.xlabel("Count")
    plt.ylabel("Lesion Type")
    plt.show()
else:
    print("⚠️ Column 'diagnostic' not found in metadata.csv")

# 7️⃣ Explore Image Paths
if 'img_id' in df.columns:
    print("\n🖼️ Sample Image IDs:")
    print(df['img_id'].head())
else:
    print("⚠️ Column 'img_id' not found in metadata.csv")


# 10️⃣ Final Verification
print("\n✅ Setup Completed Successfully! Your environment and dataset are ready for modeling.")
```
